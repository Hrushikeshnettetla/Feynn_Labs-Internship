## Segmentation and Visualization of DF1(ev_sales_by_makers_and_category_2015-2024)

```python
# Identify numeric and object columns
numeric_cols = df1.select_dtypes(include=['int64', 'float64']).columns.tolist()
object_cols = df1.select_dtypes(include=['object']).columns.tolist()
```

```python
# 1. Correlation Analysis: Heatmap of numeric columns
plt.figure(figsize=(10, 8))
corr = df1[numeric_cols].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap of Numeric Columns')
plt.show()
```

```python
#2. Frequency distribution (value counts) and count plots for each categorical column
for col in object_cols:
    print(f"\nValue Counts for {col}:")
    print(df1[col].value_counts())
    plt.figure(figsize=(8, 4))
    sns.countplot(y=col, data=df1, order=df1[col].value_counts().index)
    plt.title(f'Count Plot for {col}')
    plt.show()
```

```python
#3. GroupBy Analysis: For each object column, aggregate numeric columns (using mean)
for col in object_cols:
    print(f"\nGroupBy '{col}' - Mean of numeric columns:")
    group_mean = df1.groupby(col)[numeric_cols].mean()
    display(group_mean)
    group_mean.plot(kind='bar', figsize=(10, 6))
    plt.title(f'Mean of Numeric Columns by {col}')
    plt.ylabel('Mean')
    plt.show()
```

```python
#4. For each combination of a categorical and a numeric column, create a boxplot
for cat_col in object_cols:
    for num_col in numeric_cols:
        plt.figure(figsize=(8, 4))
        sns.boxplot(x=cat_col, y=num_col, data=df1)
        plt.title(f'{num_col} by {cat_col}')
        plt.show()
```

```python
# 3D Visualization of clusters
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111, projection='3d')

scatter = ax.scatter(df_clean['Age'], 
                     df_clean['Total Salary'], 
                     df_clean['Price'], 
                     c=df_clean['Cluster'], 
                     cmap='viridis')

ax.set_xlabel('Age')
ax.set_ylabel('Total Salary')
ax.set_zlabel('Price')
plt.title('3D Cluster Visualization')
plt.colorbar(scatter)
plt.show()

# 2D Projection using PCA
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
principal_components = pca.fit_transform(scaled_data)

plt.figure(figsize=(10, 6))
sns.scatterplot(x=principal_components[:, 0], 
                y=principal_components[:, 1], 
                hue=df_clean['Cluster'], 
                palette='viridis')
plt.title('2D Cluster Projection (PCA)')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.show()
```
